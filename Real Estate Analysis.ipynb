### Percentile Method 

```python
upper_limit = df_new['Assessed Value'].quantile(0.99)
lower_limit = df_new['Assessed Value'].quantile(0.01)
```

```python
upper_limit,lower_limit
```

```python
df_new.loc[df_new['Assessed Value'] < lower_limit]
```

### Inter Quantile Range Method 

```python
q1 = df_new['Assessed Value'].quantile(0.25)
q3 = df_new['Assessed Value'].quantile(0.75)

iqr = q3-q1
```

```python
q1,q3,iqr
```

```python
upper_limit = q3 + (1.5*iqr)
lower_limit = q1 - (1.5*iqr)

upper_limit,lower_limit
```

### Z-Score Method 

```python
upper_limit = df_new['Assessed Value'].mean() + 3*df_new['Assessed Value'].std()
lower_limit = df_new['Assessed Value'].mean() - 3*df_new['Assessed Value'].std()
```

```python
upper_limit,lower_limit
```

```python
prop_type = list(data['Property Type'].unique())
prop_type
```

```python
def get_analysis(list):
    grouping = data.groupby(['Year','Property Type'])['Sale Amount'].sum().reset_index()
    
    for item in list:
        grp = grouping.loc[grouping['Property Type'] == item]
        
        plt.figure(figsize=(10, 6))
        sns.lineplot(data=grp, x='Year', y='Sale Amount', marker='o')
        plt.title('Sales Amount Over the Years')
        plt.xlabel('Year')
        plt.ylabel(f'Property Type - {item}')
        plt.xticks(rotation=45)
        plt.tight_layout()
        plt.show()
```

```python
get_analysis(prop_type)
```

```python
resi_list = list(data['Residential Type'].unique())
resi_list
```

```python
def get_analysis1(list):
    grouping = data.groupby(['Year','Residential Type'])['Sale Amount'].sum().reset_index()
    
    for item in list:
        grp = grouping.loc[grouping['Residential Type'] == item]
        
        plt.figure(figsize=(10, 6))
        sns.lineplot(data=grp, x='Year', y='Sale Amount', marker='o')
        plt.title('Sales Amount Over the Years')
        plt.xlabel('Year')
        plt.ylabel(f'Residential Type - {item}')
        plt.xticks(rotation=45)
        plt.tight_layout()
        plt.show()
```

```python
get_analysis1(resi_list)
```

```python
def analysis2(list,n):
    grouping = data.groupby(['Year','Town'])['Sale Amount'].sum().reset_index()
    count = 0
    for item in list:
        if count == n:
            break
        grp = grouping.loc[grouping['Town'] == item]

        plt.figure(figsize=(10, 6))
        sns.lineplot(data=grp, x='Year', y='Sale Amount', marker='o')
        plt.title('Sales Amount Over the Years')
        plt.xlabel('Year')
        plt.ylabel(f'Town - {item}')
        plt.xticks(rotation=45)
        plt.tight_layout()
        plt.show()  
        count += 1
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
encoded_df = data.copy()
label_encoders = {}
```

```python
top_10_towns_list = ['Stamford',
 'Bridgeport',
 'Waterbury',
 'Norwalk',
 'New Haven',
 'Danbury',
 'West Hartford',
 'Bristol',
 'Hartford',
 'New Britain']
```

```python
encoded_df1 = encoded_df[encoded_df['Town'].isin(top_10_towns_list)]
encoded_df1
```

```python
cols = encoded_df1.columns
cols
```

```python
label_enc_cols = ['Town','Residential Type','Property Type']
```

```python
for cols in label_enc_cols:
    label_encoders[cols] = LabelEncoder()
    encoded_df1[cols] = label_encoders[cols].fit_transform(encoded_df1[cols])
```

```python
corr_matrix = encoded_df1.corr()

# Create a heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title('Correlation Heatmap')
plt.show()
```
